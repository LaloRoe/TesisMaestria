```python
import numpy as np
import matplotlib.pyplot as plt

def sistema(x, y):
    dx = y + x*(1 - x**2 - y**2)
    dy = -x + y*(1 - x**2 - y**2)
    return dx, dy

x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
dx, dy = sistema(x, y)
plt.streamplot(x, y, dx, dy, density=1.5)
plt.xlabel('x'), plt.ylabel('y')
plt.title('Campo vectorial y ciclo límite')
plt.show()
```

```python
from scipy.integrate import solve_ivp

def sistema(t, z):
    x, y = z
    dx = y + x*(1 - x**2 - y**2)
    dy = -x + y*(1 - x**2 - y**2)
    return [dx, dy]

sol = solve_ivp(sistema, [0, 100], [0.5, 0.5], method='RK45', dense_output=True)
t = np.linspace(0, 100, 1000)
x, y = sol.sol(t)
plt.plot(x, y)
```
